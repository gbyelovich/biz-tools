Categorical conversion using indicator variables allows us to do more in depth analysis and apply machine learning to make more sense of the data. Attempting to do this with basic software like excel would be extremely difficult and time consuming
FIRST WE IMPORT PANDAS AND OUR SPECIFIED COLUMNS

```python
import pandas as pd
file="ca-educational-attainment-personal-income-2008-2014.csv"
df=pd.read_csv(file,names=['Year', 'Age', 'Gender', 'Educational Attainment', 'Personal Income', 'Population Count'],skiprows=1,na_values="?")
```

```python
df.head()
```

Now we want to get a feel for object types in our columns. Need to drop first row first to remove duplicate header.

```python
df.dtypes
```

changing year to date time

```python
df.Year=pd.to_datetime(df['Year'], errors='coerce')
```

```python
df.dtypes
```

We now want to set categorical ordinal data for education level and categorical nominal indicator variable for gender to prepare our data for more meaningful analysis, ie feature representation.

```python
levelof_education=['Children under 15','No high school diploma','High school or equivalent','Some college, less than 4-yr degree',"Bachelor's degree or higher"]
df["educationlevel"]=df["Educational Attainment"].astype("category",ordered=True,categories=levelof_education).cat.codes
```

```python
gender_type=["Male","Female"]
df["Genderdummy"]=df["Gender"].astype("category",ordered=True,categories=gender_type).cat.codes
```

```python
df
```

Now lastly we want to drop education2, education attainment, and Gender columns to complete our feature representation :)

```python
df.drop(labels=["education2","Gender","Educational Attainment"],axis=1)
```
